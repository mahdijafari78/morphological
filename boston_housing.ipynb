```python
import torch
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
(x_train, y_train),(x_test, y_test) = keras.datasets.boston_housing.load_data(
    path="boston_housing.npz", test_split=0.2, seed=113
)
x_train = np.reshape(x_train, (x_train.shape[0], -1))
y_train = np.reshape(y_train, (y_train.shape[0], -1))
x_test = np.reshape(x_test, (x_test.shape[0], -1))
y_test = np.reshape(y_test, (y_test.shape[0], -1))
```

```python
model = Sequential()
model.add(Dense(100,input_dim = 13,activation='relu', name='hidd-1'))
model.add(Dense(100,activation='relu',name='hidd-2'))
model.add(Dense(1,activation='linear',name='output'))
model.compile(optimizer=keras.optimizers.Adam(lr=0.01),loss='mean_absolute_error', metrics=['mae'])
model.fit(x_train,y_train,epochs=1000)
```

```python
predict = model.predict(x_test)
for n,i in enumerate(predict):
  print(f'predict:{i[0]}')
  print(y_test[n][0])
```
